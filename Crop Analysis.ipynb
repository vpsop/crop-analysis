# Data Card

1. **N** - Ratio of Nitrogen content in soil (Numerical Datatype)
2. **P** - Ratio of Phosphorus content in soil (Numerical Datatype)
2. **K** - Ratio of Potassium content in soil (Numerical Datatype)
4. **temperature** - temperature in degree celcius (Numerical Datatype)
5. **humidity** - relative humidity of soil in % (Numerical Datatype)
6. **ph** - ph value of the soil (Numerical Datatype)
7. **rainfall** - rainfall in mm (Numerical Datatype)
8. **<u>```label```</u>** - type of crops (Target Variable) (Multiclass Classification) (Object Datatype)
